## Cleaning the IMDB Movie Dataset

Importing Pandas

```python
import pandas as pd
```

## 1. Removing any duplicates

```python
df = df.drop_duplicates()
df
```

## 2. Removing any extra whitespaces in the title and director column

```python
df["title"] = df["title"].str.strip()
df["director"] = df["director"].str.strip()
df
```

## 3. Removing any non-numeric values and brackets around the "release_year" column

```python
df["release_year"] = df["release_year"].str.strip("()II)(")
df["release_year"] = df["release_year"].str.lstrip("(")
df["release_year"] = df["release_year"].str.replace("(", "")

df["release_year"]
```

## 4. Renaming the "runtime" column into "runtime(mins)" and removing "mins" after the figures

```python
df.rename(columns = {"runtime":"runtime(mins)"}, inplace = True)

df["runtime(mins)"] = df["runtime(mins)"].str.strip("min")

## Removing extra white spaces

df["runtime(mins)"] = df["runtime(mins)"].str.strip()

df
```

## 5. Cleaning the genre column

The Genre column has numerous genres for each movie but in this case we want only one Genre, so we will be splitting the column and taking the first genre.

```python
df[["genre 1","genre 2", "genre 3"]] = df["genre"].str.split(",",2,expand = True)
df
```

Dropping the other genre columns

```python
df.drop(columns = ["genre 2", "genre 3", "genre"], inplace = True)
```

Renaming "genre 1" to "genre"

```python
df.rename(columns = {"genre 1":"genre"}, inplace = True)
```

```python
df
```

## 6. Cleaning the gross column

We want to first rename the "gross" column to "gross(M)" and then remove the $ and M from the figures

```python
df.rename(columns = {"gross": "gross(M)"}, inplace = True)

df["gross(M)"] = df["gross(M)"].str.strip("$M")
df["gross(M)"]
```

## 7. Removing the bad rows

As we can see there are numerous rows that have a gross value of 0 and a metascore value of 0. We dont want these movies.

### Removing those with a gross value of 0

```python
for x in df.index:
    if df.loc[x, "gross(M)"] == "0":
        df.drop(x, inplace = True)
        
df
```

### Removing those with a metascore of 0

```python
for x in df.index:
    if df.loc[x, "metascore"] == "0":
        df.drop(x, inplace = True)
        
df
```

## 8. Resetting the Index

```python
df = df.reset_index(drop = True)
df
```

## 9. Replacing the column names with column names that are in Proper Case

```python
cols_names = ["Title", "Director", "Release_Year", "Runtime(Mins)", "Rating", "Metascore","Gross(M)", "Genre"]
```

```python
df.columns = cols_names
```

```python
df
```

```python
df.info()
```

## 10. Changing the values in "Gross(M)" column into Float

```python
df["Gross(M)"] = df["Gross(M)"].apply(lambda x : float(x))
```
